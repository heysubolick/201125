```python
# csv 파일을 불러오기 위해 pandas library 불러오기
import pandas as pd
import matplotlib as mpl
import matplotlib.pyplot as plt
import seaborn as sns
%matplotlib inline
```

```python
mpl.rc('font', family = 'Malgun Gothic')
```

### [출처] stats.oecd.org - https://stats.oecd.org/Index.aspx?DataSetCode=FAMILY

```python
# OECD 여성의 초산 연령 데이터 불러오기

FAMILY = pd.read_csv('FAMILY.csv', encoding='utf-8')
```

```python
FAMILY
```

```python
# OECD 여성의 초산 연령 데이터에서 나라, 연도, 값(연령) 추출

OECD14 = FAMILY.loc[(FAMILY["Indicator"]=="Mean age of women at childbirth")&
                        (FAMILY['Year']>=2014),
                        ["Country","Year","Value"]]
OECD14
```

```python
# 추출한 데이터를 피벗으로 정리

OECD_p = OECD14.pivot('Country', 'Year', 'Value')
OECD_p
```

```python
# 표본 데이터가 없는 나라 드롭
OE = OECD_p.dropna(axis=0, how='any')
OE
```

```python
# OECD 국가 연도별 첫 출산 연령

fig = plt.figure(figsize=(15,15)) ## 캔버스 생성
fig.set_facecolor('white') ## 캔버스 색상 설정
plt.title('OECD국가 연도별 첫 출산 평균연령', fontsize=30)
sns.heatmap(OE, cmap='YlGnBu', annot = True, fmt='.1f')
plt.show()
```

```python
# OECD 국가 중에서 선별해 데이터를 추출

sweden=FAMILY.loc[(FAMILY["Country"]=='Sweden')&
                  (FAMILY["Indicator"]=="Mean age of women at childbirth"),
                  ["Country","Year","Value"]]
japan=FAMILY.loc[(FAMILY["Country"]=='Japan')&
                 (FAMILY["Indicator"]=="Mean age of women at childbirth"),
                 ["Country","Year","Value"]]
korea=FAMILY.loc[(FAMILY["Country"]=='Korea')&
                 (FAMILY["Indicator"]=="Mean age of women at childbirth"),
                 ["Country","Year","Value"]]
germany=FAMILY.loc[(FAMILY["Country"]=='Germany')&
                   (FAMILY["Indicator"]=="Mean age of women at childbirth"),
                   ["Country","Year","Value"]]
usa=FAMILY.loc[(FAMILY["Country"]=='United States')&
               (FAMILY["Indicator"]=="Mean age of women at childbirth"),
               ["Country","Year","Value"]]
```

```python
sweden = sweden.query('Country =="Sweden"') 
sweden_ages = sweden['Value']
sweden_years = sweden['Year']
 
japan = japan.query('Country =="Japan"') 
japan_ages = japan['Value']
japan_years = japan['Year']
 
korea = korea.query('Country =="Korea"') 
korea_ages = korea['Value']
korea_years = korea['Year']

germany = germany.query('Country =="Germany"') 
germany_ages = germany['Value']
germany_years = germany['Year']

usa = usa.query('Country =="United States"') 
usa_ages = usa['Value']
usa_years = usa['Year']
 
fig = plt.figure(figsize=(15,15)) ## 캔버스 생성
fig.set_facecolor('white') ## 캔버스 색상 설정
ax = fig.add_subplot() ## 그림 뼈대(프레임) 생성
 
ax.plot(sweden_years,sweden_ages,marker='o',markersize=10,label='스웨덴') ## 선그래프 생성
ax.plot(japan_years,japan_ages,marker='^',markersize=10,label='일본')
ax.plot(korea_years,korea_ages,marker='s',markersize=10,label='한국') 
ax.plot(germany_years,germany_ages,marker='*',markersize=10,label='독일') 
ax.plot(usa_years,usa_ages,marker='*',markersize=10,label='미국') 

plt.xticks(fontsize=20)
plt.yticks(fontsize=20)
ax.legend() ## 범례
plt.legend(fontsize=40)
plt.title('OECD 국가 연도별 첫출산 평균연령',fontsize=30) ## 타이틀 설정
plt.show()
```

```python
birthage = pd.read_csv('FAMILY.csv', encoding='utf-8')

# 쓸모없는 열(컬럼) 삭제
birthage2= birthage.drop(['Reference Period Code','Reference Period','Flag Codes','Flags','IND','PowerCode','YEAR','PowerCode Code','COU','SEX','Sex','Unit Code'], axis = 1)
birthage2

# 컬럼명을 알아보기 쉽게 수정
birthage3=birthage2.rename(columns={'Country':'나라','Indicator':'지표','Year':'연도','Unit':'단위','Value':'값'})
birthage3

birthage4 = birthage3.loc[(birthage3["지표"]=="Mean age of women at childbirth"),["나라","연도","값"]] 
birthage4

birthage5 =birthage4.groupby(['연도'], as_index=False).mean() 
birthage5
```

```python
avg_value = birthage5 ['값']
avg_year = birthage5['연도']


fig = plt.figure(figsize=(15,15)) ## 캔버스 생성
fig.set_facecolor('white') ## 캔버스 색상 설정
ax = fig.add_subplot() ## 그림 뼈대(프레임) 생성


ax.plot(avg_year,avg_value,marker='o',markersize=10,label='평균') 

plt.xticks(fontsize=20)
plt.yticks(fontsize=20)
ax.legend() ## 범례
plt.legend(fontsize=40)
plt.title('OECD 국가 평균 첫출산 평균연령',fontsize=30) ## 타이틀 설정
plt.show()
```

```python
avg_value = birthage5 ['값']
avg_year = birthage5['연도']

korea = birthage4.query('나라 =="Korea"') 
korea_value = korea['값']
korea_year = korea['연도']


fig = plt.figure(figsize=(15,15)) ## 캔버스 생성
fig.set_facecolor('white') ## 캔버스 색상 설정
ax = fig.add_subplot() ## 그림 뼈대(프레임) 생성

ax.plot(avg_year,avg_value, marker='o',markersize=10,label='OECD 평균') 
ax.plot(korea_year,korea_value, marker='o',markersize=10,label='대한민국')

plt.xticks(fontsize=20)
plt.yticks(fontsize=20)
ax.legend() ## 범례
plt.legend(fontsize=40)
plt.title('OECD 국가 평균 첫출산 평균연령',fontsize=30) ## 타이틀 설정
plt.show()
```

```python
# 고학력으로 인한 늦은 취업과 사회진출
# OECD 성인 고등교육에 대한 데이터 분석
education = pd.read_csv('성인학력.csv', encoding='utf-8')
education
```

```python
# OECD 여성의 대학교 이상 학력 평균 데이터
education2 = education.drop(['INDICATOR','SUBJECT','MEASURE','FREQUENCY','Flag Codes'], axis=1)
education3 = education2.rename(columns={'LOCATION':'나라','TIME':'연도','Value':'값'})
oecd_edu = education3.loc[(education3["나라"]=="OAVG"),["나라","연도","값"]] 
oecd_edu
```

```python
# 대한민국 여성의 대학교 이상 학력 데이터
korea_edu = education3.loc[(education3["나라"]=="KOR"),["나라","연도","값"]] 
korea_edu
```

```python
oecd_edu_value = oecd_edu['값']
oecd_edu_year = oecd_edu['연도']

korea_edu_value = korea_edu['값']
korea_edu_year = korea_edu['연도']


fig = plt.figure(figsize=(15,15)) ## 캔버스 생성
fig.set_facecolor('white') ## 캔버스 색상 설정
ax = fig.add_subplot() ## 그림 뼈대(프레임) 생성

ax.plot(oecd_edu_year ,oecd_edu_value, marker='o',markersize=10,label='OECD 평균') 
ax.plot(korea_edu_year,korea_edu_value, marker='o',markersize=10,label='대한민국')

plt.xticks(fontsize=20)
plt.yticks(fontsize=20)
ax.legend() ## 범례
plt.legend(fontsize=40)
plt.title('여성의 대학교 이상 교육 진학률',fontsize=30) ## 타이틀 설정
plt.show()
```
